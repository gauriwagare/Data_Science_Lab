## Create a 3x3 matrix

```python

A=([[2,3,1],[4,5,6],[7,8,9]])
A
```

## Determinant of a matrix

```python
detA=2*(5*9-8*6)-3*(4*9-6*7)+1*(4*8-5*7)
print(detA)
```

```python
for row in A:
    print(row)
```

## Cofactor Matrix

```python
C = [
    [
        (A[1][1]*A[2][2] - A[1][2]*A[2][1]),
       -(A[1][0]*A[2][2] - A[1][2]*A[2][0]),
        (A[1][0]*A[2][1] - A[1][1]*A[2][0])
    ],
    [
       -(A[0][1]*A[2][2] - A[0][2]*A[2][1]),
        (A[0][0]*A[2][2] - A[0][2]*A[2][0]),
       -(A[0][0]*A[2][1] - A[0][1]*A[2][0])
    ],
    [
        (A[0][1]*A[1][2] - A[0][2]*A[1][1]),
       -(A[0][0]*A[1][2] - A[0][2]*A[1][0]),
        (A[0][0]*A[1][1] - A[0][1]*A[1][0])
    ]
]
```

## Adjoint Matrix

```python
adjA = [
    [C[0][0], C[1][0], C[2][0]],
    [C[0][1], C[1][1], C[2][1]],
    [C[0][2], C[1][2], C[2][2]]
]
```

```python
print("Adjoint Matrix:")
for row in adjA:
    print(row)
```

## Inverse of a Matrix

```python
inverseA = [
    [-1/3, -19/9, 13/9],
    [ 2/3,  11/9, -8/9],
    [-1/3,  2/9, -2/9]
]

for row in inverseA:
    print(row)
```

## Linear equation solving

```python
# Coefficients
a, b = 2, 3
c, d = 4, 5

# Constants
p, q = 10, 20

# Determinant of A
detA = a*d - b*c

# Cramer's Rule
x = (p*d - b*q)/detA
y = (a*q - p*c)/detA

print("Solution using Cramer's Rule:")
print("x =", x)
print("y =", y)
```
