```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# -------------------------------
# 1. Generate Normal Distribution Data
# -------------------------------

# Parameters
mean = 50
std_dev = 10
size = 1000

# Generate random data
data = np.random.normal(mean, std_dev, size)

# -------------------------------
# 2. Descriptive Statistics
# -------------------------------

mean_value = np.mean(data)
median_value = np.median(data)
variance_value = np.var(data)
std_dev_value = np.std(data)

print("Descriptive Statistics:")
print("Mean:", mean_value)
print("Median:", median_value)
print("Variance:", variance_value)
print("Standard Deviation:", std_dev_value)

# -------------------------------
# 3. Plot Histogram with PDF
# -------------------------------

plt.hist(data, bins=10, density=True, alpha=0.6, color='g')

# Plot PDF curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'k', linewidth=2)

plt.title("Normal Distribution with PDF")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()

# -------------------------------
# 4. Coin Toss Simulation
# -------------------------------

# Simulate 1000 coin tosses
tosses = np.random.choice(['H', 'T'], size=1000)

# Count outcomes
heads = np.sum(tosses == 'H')
tails = np.sum(tosses == 'T')

# Probabilities
prob_heads = heads / 10
prob_tails = tails / 10

print("\nCoin Toss Simulation:")
print("Heads:", heads)
print("Tails:", tails)
print("Probability of Heads:", prob_heads)
print("Probability of Tails:", prob_tails)

# -------------------------------
# 5. Plot Coin Toss Result
# -------------------------------

labels = ['Heads', 'Tails']
values = [heads, tails]

plt.bar(labels, values)
plt.title("Coin Toss Results")
plt.xlabel("Outcome")
plt.ylabel("Count")
plt.show()
```
