```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

#Create a user defined dataset
data = list(map(float,input("Enter number separated by space: ").split()))
print("User defined dataset: ",data)

data_np=np.array(data)
mean_value=np.mean(data_np)

median_value=np.median(data_np)

mode_result=stats.mode(data_np)
mode_value=mode_result.mode
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Mode: {mode_value}")
```
