# Code For Factorial

```python
num=int(input("Enter the number you want: "))
fact=1
i=1
while i<=num:
  fact=fact*i
  i=i+1
print("Factorial of given number ",num,"is",fact)
```

# Code for Multiplication Table

```python
num = int(input("Enter a number to see its multiplication table: "))

print(f"Multiplication Table for {num}:")
for i in range(1, 11): # Generates table from 1 to 10
    print(f"{num} x {i} = {num * i}")
```

# Code for Checking Even And Odd Number

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print(f"{num} is an even number.")
else:
    print(f"{num} is an odd number.")
```

# Code for check palindrome

```python
num_str = input("Enter a number to check if it's a palindrome: ")

# Remove any non-digit characters and convert to string for comparison
cleaned_num_str = ''.join(filter(str.isdigit, num_str))

# Check if the cleaned number string is a palindrome
if cleaned_num_str == cleaned_num_str[::-1]:
    print(f"The number {num_str} is a palindrome.")
else:
    print(f"The number {num_str} is not a palindrome.")
```

# Code for Career Expert System

```python
def career_expert_system():
    print("=== Career Guidance Expert System ===")

    interest = input("Enter your main interest: ").lower()
    skill = input("Enter your strongest skill: ").lower()

    if interest == "hardware" and skill == "problem solving":
        career = "Computer Engineering"

    elif interest == "construction" and skill == "design":
        career = "Civil Engineering"

    elif interest == "machines" and skill == "mechanics":
        career = "Mechanical Engineering"

    elif interest == "data" and skill == "probabilty & statics":
        career = "AI & Data Science (AIDS Engineering)"

    elif interest == "programming" and skill == "logic":
        career = "Software Engineering"

    else:
        career = "No suitable career found based on the given inputs."

    print("\nRecommended Career:", career)


career_expert_system()

```
