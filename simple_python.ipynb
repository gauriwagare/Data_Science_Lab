```python
a = 10
b = 5

# Addition
sum_result = a + b
print(f"Addition: {a} + {b} = {sum_result}")

# Subtraction
diff_result = a - b
print(f"Subtraction: {a} - {b} = {diff_result}")

# Multiplication
prod_result = a * b
print(f"Multiplication: {a} * {b} = {prod_result}")

# Division
div_result = a / b
print(f"Division: {a} / {b} = {div_result}")

# Floor Division (integer division)
floor_div_result = a // b
print(f"Floor Division: {a} // {b} = {floor_div_result}")

# Modulus (remainder)
mod_result = a % b
print(f"Modulus: {a} % {b} = {mod_result}")

# Exponentiation
pow_result = a ** 2
print(f"Exponentiation: {a} ** 2 = {pow_result}")
```

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Example usage:
number = 5
print(f"The factorial of {number} is: {factorial(number)}")

number = 0
print(f"The factorial of {number} is: {factorial(number)}")

number = -3
print(f"The factorial of {number} is: {factorial(number)}")
```

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
num1 = 7
print(check_odd_even(num1))

num2 = 10
print(check_odd_even(num2))

num3 = 0
print(check_odd_even(num3))
```

```python
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the original string with its reversed version
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
num1 = 121
print(is_palindrome(num1))

num2 = 12321
print(is_palindrome(num2))

num3 = 12345
print(is_palindrome(num3))

num4 = 55
print(is_palindrome(num4))
```

## Design Knowledge Base (Rules)

### Subtask:
Define the rules that the expert system will use to make decisions. This will include criteria like cuisine type, price range, and the corresponding recommendations.


**Reasoning**:
The user wants to define rules for an expert system by creating a knowledge base. I will use a Python code block to create a list of dictionaries, where each dictionary represents a rule with 'cuisine', 'price_range', and 'recommendation' keys, as instructed.



```python
knowledge_base = [
    {
        'cuisine': 'Italian',
        'price_range': 'moderate',
        'recommendation': 'Pasta Paradise'
    },
    {
        'cuisine': 'Mexican',
        'price_range': 'cheap',
        'recommendation': 'Taco Fiesta'
    },
    {
        'cuisine': 'Indian',
        'price_range': 'expensive',
        'recommendation': 'Spice Route'
    },
    {
        'cuisine': 'American',
        'price_range': 'moderate',
        'recommendation': 'Burger Joint'
    },
    {
        'cuisine': 'Italian',
        'price_range': 'expensive',
        'recommendation': 'Fine Italian Dining'
    },
    {
        'cuisine': 'Mexican',
        'price_range': 'moderate',
        'recommendation': 'Authentic Mexican Grill'
    },
    {
        'cuisine': 'Indian',
        'price_range': 'cheap',
        'recommendation': 'Curry House'
    }
]

print("Knowledge Base (Rules) created successfully:")
for rule in knowledge_base:
    print(rule)
```

## Implement Inference Engine

### Subtask:
Develop a function that takes user preferences (cuisine and price range) as input and uses the defined knowledge base to find and return a matching restaurant recommendation.

**Reasoning**:
Now that the knowledge base is defined, I will implement the inference engine as a Python function. This function will take user input for cuisine and price range and iterate through the `knowledge_base` to find a matching recommendation.



```python
def get_recommendation(cuisine, price_range, kb):
    """
    Recommends a restaurant based on cuisine and price range using the knowledge base.
    """
    for rule in kb:
        if rule['cuisine'].lower() == cuisine.lower() and \
           rule['price_range'].lower() == price_range.lower():
            return f"Recommended restaurant: {rule['recommendation']}"
    return "Sorry, no recommendation found for your preferences."

# Example usage of the inference engine:
user_cuisine = "Italian"
user_price = "moderate"
recommendation = get_recommendation(user_cuisine, user_price, knowledge_base)
print(f"User Preferences: Cuisine = {user_cuisine}, Price Range = {user_price}")
print(recommendation)

user_cuisine = "Mexican"
user_price = "expensive"
recommendation = get_recommendation(user_cuisine, user_price, knowledge_base)
print(f"\nUser Preferences: Cuisine = {user_cuisine}, Price Range = {user_price}")
print(recommendation)

user_cuisine = "Indian"
user_price = "moderate"
recommendation = get_recommendation(user_cuisine, user_price, knowledge_base)
print(f"\nUser Preferences: Cuisine = {user_cuisine}, Price Range = {user_price}")
print(recommendation)
```

```python
print("\n--- Restaurant Recommendation System ---")
user_cuisine_input = input("Enter preferred cuisine (Italian, Mexican, Indian, American): ")
user_price_input = input("Enter preferred price range (cheap, moderate, expensive): ")

final_recommendation = get_recommendation(user_cuisine_input, user_price_input, knowledge_base)
print(f"\nYour choice: Cuisine = {user_cuisine_input}, Price Range = {user_price_input}")
print(final_recommendation)
```
