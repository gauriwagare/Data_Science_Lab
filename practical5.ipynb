```python
import pandas as pd

# Load dataset
df = pd.read_csv("tested.csv")

# Display dataset
print(df.head())


# -----------------------
# Filtering Data
# -----------------------

# Passengers older than 30
age_filter = df[df['Age'] > 30]
print("Passengers older than 30:")
print(age_filter.head())

# Female passengers
female_passengers = df[df['Sex'] == 'female']
print("Female passengers:")
print(female_passengers.head())


# -----------------------
# Grouping Data
# -----------------------

# Group by Passenger Class
fare_by_class = df.groupby('Pclass')['Fare'].mean()
print("Average Fare by Passenger Class:")
print(fare_by_class)

# Survival rate by gender
survival_by_gender = df.groupby('Sex')['Survived'].mean()
print("Survival Rate by Gender:")
print(survival_by_gender)


# -----------------------
# Aggregate Statistics
# -----------------------

print("Mean Age:", df['Age'].mean())
print("Total Fare:", df['Fare'].sum())
print("Passenger Count:", df['PassengerId'].count())
print("Minimum Fare:", df['Fare'].min())
print("Maximum Fare:", df['Fare'].max())
```
